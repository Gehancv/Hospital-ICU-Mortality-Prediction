```python
import pandas as pd
import numpy as np

data= pd.read_csv(r"C:\Users\saras\Documents\School\STA 6704- Data Mining 2\widsdatathon2020\training_v2.csv")
data.head()
```

```python
pd.set_option('display.max_columns', None)
distri_original=data.describe()
distri_original
```

```python
pd.set_option('display.max_rows', None)
data.isnull().sum()
#To be noted that the target variable hospital_death do not have any missing values
```

```python
#Dropping columns with more than 15% missing values
data_new=data[data.columns[data.isnull().sum()<(0.15*data.shape[0])]]
data_new.shape
```

```python
data_new.isnull().sum()
```

```python
#Dropping the rows with null values and then check the distribution
data_new=data_new.dropna(axis=0)
data_new.shape
```

```python
distri_new=data_new.describe()
distri_new
```

```python
#Checking the percentage change in the mean and median after removing the rows with null values
a = np.intersect1d(distri_new.columns, distri_original.columns).tolist()
result=pd.DataFrame()
for i in a:
    result[i+'_percent']= ((distri_new[i]/distri_original[i])-1)*100
result.loc[['mean','50%'],:]
```
